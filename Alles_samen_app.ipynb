```python
fig, ax = plt.subplots(nrows=1, ncols=1, dpi=150, figsize=(7, 3))

def tijd(h, g,alpha,mu):
    """tijd"""
    return np.sqrt((2*h)/(g*(np.sin(alpha) - mu * np.cos(alpha))))



#Vervolgens wordt er een initiële fit gemaakt van de data door waarden voor de parameters te gokken, 
#zodat de fit goed past op de data. 
#Na enig puzzelwerk, werden de volgende waarden voor de paramters gekozen die een goede fit lijken te bekomen:

def plot(ax, x, y):
    """Plotten functie Amplitude"""
    ax.plot(x, y, color = 'y', linewidth=2, label = 'Hout ; ' r'$R^{2} = $' + str(0.98))
    
def plot2(ax, x, y):
    """Plotten functie Amplitude"""
    ax.plot(x, y, color = 'r', linewidth=2, label = 'Leer ; ' r'$R^{2} = $' + str(0.92))

g = 9.81
alpha1 = 50*np.pi/180
alpha2 = 40*np.pi/180
mu2 = 0.46467103
mu1 = 0.54887774


z = np.linspace(0.1, 1.1, num=5000)

plot(ax, z, tijd(z,g, alpha1, mu1))
plot2(ax, z, tijd(z,g, alpha1, mu2))
ax.grid(True, which = 'both')


ax.set_title('Vergelijking van de data t(h) voor verschillende materialen op helling bij ' r'$\alpha$ = ' + str(50) + '$^\circ$')
ax.grid(True, which = 'both') #Voegt rijen en kolommen toe, zodat de datapunten beter vergeleken en afgelezen kunnen worden.
plt.errorbar(hoogte1pop, tijd1_hout,s_tijd1_hout, shoogte, label = 'hout', fmt = 'd ', c='y', capsize=3, elinewidth=0.5, capthick=0.5)
plt.errorbar(hoogte1, tijd1_leer,s_tijd1_leer, shoogte, label = 'leer', fmt = 'd ', c='r', capsize=3, elinewidth=0.5, capthick=0.5)
plt.ylabel('t [s]', size=10)
plt.xlabel('h [m]', size=10)
plt.legend(fontsize=8)
ax.set_xlim([0.1, 1.1])

plt.tick_params(axis='both', which='major', labelsize=10)

fig.savefig('tijd_versus_hoogte_app_verschillende_meterialen_alpha_50', bbox_inches = 'tight')


plt.show()
```
